```python
import os
import numpy as np
import matplotlib.pyplot as plt
from cp_utils.cp_dir_file_ops import define_env

if define_env() == 'SERVER':
    dir = '/home/neo/PycharmProjects/Course1/subprojects/sp02_omniiseg_fgn_mmdet/MNISTISEG_N3_K3'
elif define_env() == 'PC':
    dir = 'C:/Users/Art/PycharmProjects/Course1/subprojects/sp02_omniiseg_fgn_mmdet/MNISTISEG_N3_K3'
else:
    raise NotImplementedError

tag = 'MNISTISEG_val_FT_Ignore_bbox_mAP_base_'
```

```python
sp0 = f'2022-02-13_09-37-40_Train-3-3-Scratch-MNISTISEG-Train LR 0.001 WD 1e-05 Adagrad-tag-{tag}.csv'
sp1 = f'2022-02-13_09-41-18_Train-3-3-Scratch-MNISTISEG-Train LR 0.005 WD 1e-05 Adagrad-tag-{tag}.csv'
sp2 = f'2022-02-13_17-33-19_Train-3-3-Scratch-MNISTISEG-Train LR 0.01 WD 1e-05 Adagrad-tag-{tag}.csv'
sp3 = f'2022-02-13_13-43-29_Train-3-3-Scratch-MNISTISEG-Train LR 0.05 WD 1e-05 Adagrad-tag-{tag}.csv'

fp0 = os.path.join(dir, sp0)
fp1 = os.path.join(dir, sp1)
fp2 = os.path.join(dir, sp2)
fp3 = os.path.join(dir, sp3)

# Wall time, Step, Value
data0 = np.around(np.loadtxt(fp0, delimiter=',', skiprows=1)[:, 2], decimals=3)
data1 = np.around(np.loadtxt(fp1, delimiter=',', skiprows=1)[:, 2], decimals=3)
data2 = np.around(np.loadtxt(fp2, delimiter=',', skiprows=1)[:, 2], decimals=3)
data3 = np.around(np.loadtxt(fp3, delimiter=',', skiprows=1)[:, 2], decimals=3)

fig = plt.figure(figsize=(5, 4), dpi=100, facecolor='white')

plt.ylabel('  ')
accuracy = np.arange(11, dtype=np.int32) / 10
plt.yticks(accuracy)
plt.ylim(0.1, 1.05)

plt.xlabel('')
epochs = np.arange(10, dtype=np.int32) + 1
plt.xticks(epochs)

plt.plot(epochs, data0, label=r'$\eta = 0.001 $')
plt.plot(epochs, data1, label=r'$\eta = 0.005 $')
plt.plot(epochs, data2, label=r'$\eta = 0.01 $')
plt.plot(epochs, data3, label=r'$\eta = 0.05 $')

plt.grid(True)
plt.legend(loc='upper left')
plt.tight_layout()
# plt.title(f'Experiments with different learning rates $\gamma$')
plt.savefig('MNISTISEG_N3_K3_Experiments_LR.png', facecolor=fig.get_facecolor(), edgecolor='none', transparent=False)
plt.show()

# Some
```
